```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df_mcplant = pd.read_csv("mcplant10k_date_with_score.csv")
df_mcplant["Label"] = df_mcplant["Score"].apply(lambda x: "Positive" if x >= 0.5 else "Negative")
#df_mcplant["Label"].value_counts()
# McPlant Sentiment Breakdown

plt.style.use('ggplot')
plt.figure(figsize=(30,30))

labels = 'Positive', 'Negative'
sizes = [8033, 1967]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'size': 15})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.legend(loc=1,fontsize=10)
plt.title("McPlant Sentiment Breakdown",fontweight="bold" ,fontsize="15")



fig1.savefig('McPlant_SentimentBreakdown.png', dpi=100)
```
